```python
import pickle

'''
------------------------------------------------------------------------------------------------------
Example of loading train/val/test data
'''

def print_img_gt(data, num=10, distinct_images=True):
    count = 0
    image_ids = data['image_id']
    gt_sents = data['gold_sents']
    printed = []
    for img, gt in zip(image_ids, gt_sents):
        if distinct_images and img in printed:
            continue
        print("{}\t{}".format(img, gt))
        printed.append(img)
        count += 1
        if count >= num:
            break
```

A **PKL** file is a file created by pickle, a Python module that enabless objects to be serialized to files on disk and deserialized back into the program at runtime. It contains a byte stream that represents the objects.

The process of serialization is called "pickling," and deserialization is called "unpickling." A PKL file is pickled to save space when being stored or transferred over a network then is unpickled and loaded back into program memory during runtime. The PKL file is created using Python pickle and the **dump()** method and is loaded using Python pickle and the **load()** method.

```python
def col_summary(data):
    print('- count:',len(data))
    print('- count distinct:',len(set(data)))
    print('- dtype: str')
    length = data.str.len()
    print('- maximum length:',max(length))
    print('- minimum length:',min(length))
```

```python
def list_dim(data):
    dim = dict()
    for item in data['sents_sim_score']:
        if not len(item) in dim:
            dim[len(item)] = 1
        else:
            dim[len(item)]+= 1
    print(dim)
```

This module implements word vectors and their similarity look-ups.

Since trained word vectors are independent from the way they were trained (Word2Vec, FastText, WordRank, VarEmbed etc), they can be represented by a standalone structure, as implemented in this module.

The structure is called KeyedVectors and is essentially a mapping between entities and vectors. Each entity is identified by its string id, so this is a mapping between **{str => 1D numpy array}**.

The entity typically corresponds to a word (so the mapping maps words to 1D vectors), but for some models, the key can also correspond to a document, a graph node etc. To generalize over different use-cases, this module calls the keys **entities**. Each entity is always represented by its string id, no matter whether the entity is a word, a document or a graph node.

```python
'''
------------------------------------------------------------------------------------------------------
load the pretrained word embeddings 
'''
from gensim.models import KeyedVectors
we_file = 'glove.840B.300d.top10k.word2vec'
embeddings = KeyedVectors.load_word2vec_format(we_file)

# getting the embeddings for the word "pizza"
pizza = embeddings['pizza']
```
